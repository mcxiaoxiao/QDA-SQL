```python
import os
import json
import sqlparse
import matplotlib.pyplot as plt
from collections import defaultdict

# sql_depth ÂáΩÊï∞ÔºåÁî®‰∫éËÆ°ÁÆó SQL Êü•ËØ¢ÁöÑÊ∑±Â∫¶
def sql_depth(sql):
    def get_depth(node):
        if not isinstance(node, sqlparse.sql.TokenList):
            return 0
        else:
            return 1 + max((get_depth(child) for child in node), default=0)
    parsed = sqlparse.parse(sql)[0]
    depth = get_depth(parsed)
    return depth

# ÈÄíÂΩíÂú∞ÊêúÁ¥¢Âê´Êúâ query ÁöÑ‰∏™‰ΩìÂπ∂ËÆ°ÁÆó SQL Ê∑±Â∫¶
def search_and_calculate_depth(data, depths):
    if isinstance(data, dict):
        if 'query' in data:
            sql = data['query']
            depth = sql_depth(sql)
            depths[depth] += 1
        for value in data.values():
            search_and_calculate_depth(value, depths)
    elif isinstance(data, list):
        for item in data:
            search_and_calculate_depth(item, depths)

# ÂàùÂßãÂåñÊ∑±Â∫¶ËÆ°Êï∞Âô®
depths = defaultdict(int)

# ÈÅçÂéÜÁõÆÂΩï‰∏≠ÁöÑÊâÄÊúâ JSON Êñá‰ª∂
directory = "c_outputs/geminiÁöÑ0-100"
for filename in os.listdir(directory):
    if filename.endswith(".json"):
        file_path = os.path.join(directory, filename)
        with open(file_path, 'r') as file:
            data = json.load(file)
            search_and_calculate_depth(data, depths)

# Ê£ÄÊü•ÊòØÂê¶ÊúâÊ∑±Â∫¶‰ø°ÊÅØÂπ∂ÁªòÂõæ
if depths:  # Â¶ÇÊûú depths ‰∏ç‰∏∫Á©∫
    plt.figure(figsize=(10, 6))
    plt.bar(depths.keys(), depths.values())
    plt.title('Distribution of SQL Depths in JSON Files')
    plt.xlabel('Depth')
    plt.ylabel('Frequency')
    plt.xticks(range(min(depths.keys()), max(depths.keys()) + 1))
    plt.grid(axis='y', linestyle='--')
    plt.show()
else:
    print("No SQL queries found in the generated QAs.")
```

```python
import os
import json
import sqlparse
import matplotlib.pyplot as plt
from collections import defaultdict



# sql_depth ÂáΩÊï∞ÔºåÁî®‰∫éËÆ°ÁÆó SQL Êü•ËØ¢ÁöÑÊ∑±Â∫¶
def sql_depth(sql):
    def get_depth(node):
        if not isinstance(node, sqlparse.sql.TokenList):
            return 0
        else:
            return 1 + max((get_depth(child) for child in node), default=0)
    parsed = sqlparse.parse(sql)[0]
    depth = get_depth(parsed)
    return depth

# ‰øÆÊîπÂêéÁöÑÈÄíÂΩíÊêúÁ¥¢ÂáΩÊï∞Ôºå‰∏ìÈó®Áî®‰∫éÂ§ÑÁêÜturns‰∏≠ÁöÑquery
def search_and_calculate_depth_in_turns(data, depths):
    if isinstance(data, list):
        for item in data:
            if 'query' in item:  # ÂÅáËÆæÊØè‰∏™turnÊòØ‰∏Ä‰∏™Â≠óÂÖ∏Ôºå‰∏îqueryÁõ¥Êé•Âú®turnÂÜÖ
                sql = item['query']
                depth = sql_depth(sql)
                depths[depth] += 1

# ÂàùÂßãÂåñÊ∑±Â∫¶ËÆ°Êï∞Âô®
depths = defaultdict(int)

# ÊåáÂÆöJSONÊñá‰ª∂Ë∑ØÂæÑ
file_path = "c_outputs/gemini_test_optimized_filtered.json"
with open(file_path, 'r') as file:
    data = json.load(file)
    # ÂÅáËÆæJSONÁªìÊûÑÊòØ‰∏Ä‰∏™ÂàóË°®ÔºåÊØè‰∏™ÂÖÉÁ¥†ÂåÖÂê´‰∏Ä‰∏™turnsÂ≠óÊÆµ
    for entry in data:
        search_and_calculate_depth_in_turns(entry['turns'], depths)

# Ê£ÄÊü•ÊòØÂê¶ÊúâÊ∑±Â∫¶‰ø°ÊÅØÂπ∂ÁªòÂõæ
if depths:  # Â¶ÇÊûú depths ‰∏ç‰∏∫Á©∫
    plt.figure(figsize=(10, 6))
    plt.bar(depths.keys(), depths.values(), color='green')
    plt.title('Distribution of SQL Depths in JSON Files')
    plt.xlabel('Depth')
    plt.ylabel('Frequency')
    plt.xticks(range(min(depths.keys()), max(depths.keys()) + 1))
    plt.grid(axis='y', linestyle='--')
    plt.show()
else:
    print("No SQL queries found in the provided data.")
```

```python
import pandas as pd
import json
import sqlparse
import matplotlib.pyplot as plt
import numpy as np

# ‰Ω†‰πãÂâçÂÆö‰πâÁöÑ sql_depth ÂáΩÊï∞‰øùÊåÅ‰∏çÂèò

# ËØªÂèñCSVÊñá‰ª∂Âπ∂ËÆ°ÁÆóSQLÊ∑±Â∫¶ÂàÜÂ∏É
df = pd.read_csv('goals_of_BIRD.csv')
df['sql_depth'] = df['goal_sql'].apply(sql_depth)
original_depth_distribution = df['sql_depth'].value_counts().sort_index()

# ËØªÂèñËøáÊª§ÂêéÁöÑJSONÊñá‰ª∂Âπ∂ËÆ°ÁÆóSQLÊ∑±Â∫¶ÂàÜÂ∏É
def calculate_depth_from_json(file_path):
    depths = []
    with open(file_path, 'r') as file:
        data = json.load(file)
        for entry in data:
            for turn in entry['turns']:
                if 'query' in turn:
                    depth = sql_depth(turn['query'])
                    depths.append(depth)
    return pd.Series(depths).value_counts().sort_index()

filtered_depth_distribution = calculate_depth_from_json('c_outputs/gemini_test_optimized_filtered.json')

# ÂàõÂª∫‰∏Ä‰∏™DataFrameÊù•Êï¥Âêà‰∏§‰∏™ÂàÜÂ∏É
combined_distribution = pd.DataFrame({
    'Original': original_depth_distribution,
    'Filtered': filtered_depth_distribution
}).fillna(0)  # Â°´ÂÖÖNaNÂÄº‰∏∫0

# ÁªòÂà∂ÂèåËâ≤Êü±Áä∂Âõæ
combined_distribution.plot(kind='bar', figsize=(10, 6), color=['#7ABBDB', '#84BA42'])
plt.title('Comparison of SQL Depth Distribution')
plt.xlabel('Depth')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')

plt.show()
```
